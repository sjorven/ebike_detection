```python
import sqlite3 as sl
import pandas as pd
from matplotlib import pyplot as plt
```

# Lineplots for one activity

```python
conn = sl.connect("cycling.db")
 
sql = """
    select velocity_smooth, heartrate, grade_smooth, time, activity_id 
    from data_raw
    where activity_id = 6916950829
"""
 
data = pd.read_sql(sql, conn)
data.velocity_smooth = data.velocity_smooth.apply(lambda x: x*3.6)

plt.subplot(311)
plt.plot(data.time, data.velocity_smooth)
plt.xlabel("time (s)")
plt.ylabel("velocity (km/h)")
plt.grid(1)

plt.subplot(312)
plt.plot(data.time, data.heartrate)
plt.xlabel("time (s)")
plt.ylabel("heartrate (bpm)")
plt.grid(1)

plt.subplot(313)
plt.plot(data.time, data.grade_smooth)
plt.xlabel("time (s)")
plt.ylabel("gradient (%)")
plt.grid(1)

plt.subplots_adjust(left=0.10, right=0.95, hspace=0.8)
plt.show()
```

```python
plt.plot(data.time, data.heartrate)
plt.xlabel("time (s)")
plt.ylabel("heartrate (bpm)")
plt.grid(1)
plt.show()
```

# Histograms for comparison of e-bike and not e-bike

```python
conn = sl.connect("cycling.db")
ebike_query = """
    select velocity_smooth, heartrate, grade_smooth, time, activity_id 
    from data_raw
    where label = 1
"""
not_ebike_query = """
    select velocity_smooth, heartrate, grade_smooth, time, activity_id 
    from data_raw
    where label = 0
"""
ebike_data = pd.read_sql(ebike_query, conn)
not_ebike_data = pd.read_sql(not_ebike_query, conn)
ebike_data.velocity_smooth = ebike_data.velocity_smooth.apply(lambda x: x*3.6)
not_ebike_data.velocity_smooth = not_ebike_data.velocity_smooth.apply(lambda x: x*3.6)
```

```python
plt.hist(ebike_data.velocity_smooth, bins=30, alpha=0.5,color="blue" ,label='e-bike')
plt.hist(not_ebike_data.velocity_smooth, bins=40, alpha=0.5, color = 'green', label='bike')
plt.xlim(0,70)
plt.xlabel('Velocity in km/h')
plt.ylabel('Frequency')
plt.legend()
plt.grid(1)
plt.show()
```

```python
plt.hist(ebike_data.heartrate, bins=30, alpha=0.5,color = 'blue', label='e-bike')
plt.hist(not_ebike_data.heartrate, bins=30, alpha=0.5,color= "green" ,label='bike')
plt.xlabel('Heartrate in bpm')
plt.ylabel('Frequency')
plt.legend()
plt.grid(1)
plt.show()
```

```python
plt.hist(ebike_data.grade_smooth, bins=25, alpha=0.5,color="blue", label='e-bike')
plt.hist(not_ebike_data.grade_smooth, bins=25, alpha=0.5, color = "green",label='bike')
plt.xlabel('Gradient in percent')
plt.ylabel('Frequency')
plt.legend()
plt.grid(1)
plt.show()
```
